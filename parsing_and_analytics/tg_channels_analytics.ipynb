```python
import pandas as pd
pd.set_option('display.max_colwidth', 500)

df = pd.read_csv('telegram_messages.csv')

df['content'] = df['content'].str.lower()  # –ø—Ä–∏–≤–æ–¥–∏–º –∫–æ–Ω—Ç–µ–Ω—Ç –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É

# –û–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–π —à–∞–±–ª–æ–Ω –¥–ª—è –∏–º–µ–Ω–∏ —Å —É—á–µ—Ç–æ–º —Ä—É—Å—Å–∫–æ–≥–æ –∏ –∞–Ω–≥–ª–∏–π—Å–∫–æ–≥–æ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ (–∏–º—è –æ–ø—Ü–∏–æ–Ω–∞–ª—å–Ω–æ)
name_pattern = r"((–ø–∞–≤–µ–ª(?:–∞)?\s+)?–¥—É—Ä–æ–≤(?:–∞)?|(pavel\s+)?durov)"
# –û–±–Ω–æ–≤–ª–µ–Ω–Ω—ã–π —à–∞–±–ª–æ–Ω –¥–ª—è –∫–ª—é—á–µ–≤—ã—Ö —Å–ª–æ–≤ —Å –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ–º telegram
keywords_pattern = r"(—Ñ—Ä–∞–Ω—Ü\w*|france\w*|–¥—É–±–∞–π\w*|dubai\w*|—Å—É–¥\w*|court\w*|ton\w*|telegram\w*)"

filtered = df[
    df['content'].str.contains(name_pattern, regex=True, na=False) &
    df['content'].str.contains(keywords_pattern, regex=True, na=False)
]

# –°–æ–∑–¥–∞–¥–∏–º –∫–æ–ø–∏—é –¥–ª—è –∏–∑–±–µ–∂–∞–Ω–∏—è –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–π SettingWithCopy
filtered = filtered.copy()

# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–æ–Ω–∫—É —Å —Ä–∞–Ω–≥–æ–º –ø–æ –¥–∞—Ç–µ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–∞–ª–∞
filtered['channel_rank'] = filtered.groupby('channel')['date'].rank(method='dense', ascending=True)

# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –æ–±—â–∏–º —Ä–∞–Ω–≥–æ–º –ø–æ –¥–∞—Ç–µ
filtered['overall_rank'] = filtered['date'].rank(method='dense', ascending=True)
```

```python
filtered[filtered['overall_rank'] < 21].sort_values(by='overall_rank').head(20)
```

```python
news_pattern = r"(?i)(|sec(?:\b|\.?e\.?c\.?)|securities?|ripple|—Ä–∏–ø–ø–ª|xrp|appeal|–∞–ø–µ–ª–ª—è—Ü–∏\w*|brad\s+garlinghouse|–±—Ä—ç–¥\s+–≥–∞—Ä–ª–∏–Ω–≥—Ö–∞—É—Å)"

filtered = df[df['content'].str.contains(news_pattern, regex=True, na=False)]

# –°–æ–∑–¥–∞–¥–∏–º –∫–æ–ø–∏—é –¥–ª—è –∏–∑–±–µ–∂–∞–Ω–∏—è –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–π SettingWithCopy
filtered = filtered.copy()

# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–æ–Ω–∫—É —Å —Ä–∞–Ω–≥–æ–º –ø–æ –¥–∞—Ç–µ –≤–Ω—É—Ç—Ä–∏ –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–∞–ª–∞
filtered['channel_rank'] = filtered.groupby('channel')['date'].rank(method='dense', ascending=True)

# –î–æ–±–∞–≤–ª—è–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –æ–±—â–∏–º —Ä–∞–Ω–≥–æ–º –ø–æ –¥–∞—Ç–µ
filtered['overall_rank'] = filtered['date'].rank(method='dense', ascending=True)
```
