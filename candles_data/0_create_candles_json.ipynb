```python
plot_df = pd.read_parquet('03_05_1m.parquet')
plot_df = plot_df.loc[(plot_df.ticker == 'SOL/USDT') & (plot_df.datetime.between('2025-03-03 18:00:00', '2025-03-04 21:59:59'))]    
```

```python
import plotly.graph_objs as go

fig = go.Figure(
    data=[go.Candlestick(
        x=plot_df['datetime'],
        open=plot_df['open'],
        high=plot_df['high'],
        low=plot_df['low'],
        close=plot_df['close'],
        name='OHLC'
    )]
)

fig.update_layout(
    title='BTC/USDT Свечной график',
    xaxis_title='Время',
    yaxis_title='Цена (USDT)'
)

fig.show()
```

```python
def duplicate_with_symbols(data, symbols):
    result = []
    for item in data:
        for symbol in symbols:
            new_item = item.copy()  # Create a copy of the original dictionary
            new_item['s'] = symbol  # Replace the symbol
            result.append(new_item)
    return result

# List of symbols you want to include
symbols = ['solusdt', 'ethusdt', 'mntusdt', 'xrpusdt', 'adausdt', 'dogeusdt']

# Example usage (assuming your original data is stored in a variable called 'data')
data = duplicate_with_symbols(data, symbols)
```
